```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns", None) 
pd.set_option("display.width", None)       
```

```python
file_path = '---'  
ghg_emissions = pd.read_csv(file_path)
```

```python
print("\n Info on DataFrame :")
ghg_emissions.info()  
```

```python
ghg_emissions.drop(columns=['ISO2'], inplace=True)
```

```python
first_50_rows = ghg_emissions.head(50)  
print("\n First 50 rows of Dataframe")
print(first_50_rows)
```

```python
total_emissions = ghg_emissions.groupby('Industry')['F2022'].sum().reset_index()
```

```python
plt.figure(figsize=(14, 8))  
sns.barplot(x='Industry', y='F2022', hue='Gas_Type', data=ghg_emissions) 
plt.title('Greenhouse Gas Emissions by Sector in 2022')  
plt.xlabel('Industry')  
plt.ylabel('Emissions (units)')  
plt.xticks(rotation=45, ha='right', fontsize=6)  
plt.legend(title='Gas Type')    
plt.tight_layout()  
plt.show()
```
