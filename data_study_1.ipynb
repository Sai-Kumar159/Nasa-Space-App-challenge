```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns", None)  
pd.set_option("display.width", None)     
```

```python
file_path = '---'
ghg_emissions = pd.read_csv(file_path)
```

```python
print("\n Info on DataFrame :")
ghg_emissions.info() 
```

```python
ghg_emissions.drop(columns=['ISO2'], inplace=True) 
```

```python
first_50_rows = ghg_emissions.head(50) 
print("\n First 50 rows of Dataframe")
print(first_50_rows) 
```

```python
total_emissions = ghg_emissions.groupby('Industry')['F2022'].sum().reset_index()
```

```python
plt.figure(figsize=(14, 7))     
plt.bar(total_emissions['Industry'], total_emissions['F2022'], color='green')   
plt.title('Total Greenhouse Gas Emissions by Industry in 2022')     
plt.xlabel('Industry')  
plt.ylabel('Total Emissions (units)')   
plt.xticks(rotation=90)     
plt.tight_layout()  
plt.show()  
```
